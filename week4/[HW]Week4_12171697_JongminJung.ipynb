```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K = 10
rand_data = np.random.rand(2, 100)
x = np.random.rand(2, 1)
```

```python
dataLen = rand_data.shape[1]
if dataLen < K:
    K = dataLen
```

# 유클리디안 거리

```python
distance = np.sqrt(((rand_data - x)**2).sum(axis=0))
idx = np.argsort(distance)
neighborIndex = idx[:K]
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label="Data")
plt.plot(x[0, 0], x[1, 0], "or", label="Query")
plt.plot(rand_data[0, neighborIndex], rand_data[1, neighborIndex], 'o', markeredgewidth=0.5, label='KNN')
plt.legend(loc='upper right')
plt.show()
```

# 맨해튼 거리

```python
distance = (np.abs((rand_data - x)).sum(axis=0))
idx = np.argsort(distance)
neighborIndex = idx[:K]
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label="Data")
plt.plot(x[0, 0], x[1, 0], "or", label="Query")
plt.plot(rand_data[0, neighborIndex], rand_data[1, neighborIndex], 'o', markeredgewidth=0.5, label='KNN')
plt.legend(loc='upper right')
plt.show()
```

# 코사인 거리

```python
distance = []
Y = [x[0][0], x[1][0]]
for i in range(dataLen):
    X = [rand_data[0][i], rand_data[1][i]]
    distance.append(1 - (np.dot(X, Y) / (np.linalg.norm(X) * (np.linalg.norm(Y)))))

idx = np.argsort(distance)
neighborIndex = idx[:K]
```

```python
plt.plot(rand_data[0, :], rand_data[1, :], "ok", label="Data")
plt.plot(x[0, 0], x[1, 0], "or", label="Query")
plt.plot(rand_data[0, neighborIndex], rand_data[1, neighborIndex], 'o', markeredgewidth=0.5, label='KNN')
plt.legend(loc='upper right')
plt.show()
```
